```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#pip install scikit-image
from skimage import data
img = data.chelsea()
plt.imshow(img)
```

```python
img.shape
```

```python
#Normalisation et applatissement de l'image
img_data = (img/255.0).reshape(-1,3)
img_data.shape
```

```python
from sklearn.cluster import KMeans
```

```python
KMeans(n_clusters=1)
```

```python
kmeans = KMeans(n_clusters=3).fit(img_data)
```

```python
k_colors= kmeans.cluster_centers_[kmeans.predict(img_data)]
k_colors
```

```python
kmeans.predict(img_data)
```

```python
kmeans.cluster_centers_
```

```python
img.shape
```

```python
k_colors.shape
```

```python
k_img = np.reshape(k_colors, (img.shape))
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize = (12, 4.5))
fig.suptitle('K-Means Image Compression',fontsize=20)
ax1.set_title('Compressed')
ax1.imshow(k_img)

ax2.set_title('Orignal (16,777,216 colors)')
ax2.imshow(img/255)
plt.subplots_adjust(top=0.85)
plt.show()
```

```python
kmeans = KMeans(n_clusters=3).fit(img_data)

k_colors= kmeans.cluster_centers_[kmeans.predict(img_data)]

k_img = np.reshape(k_colors, (img.shape))

fig, (ax1,ax2) = plt.subplots(1,2, figsize = (12, 4.5))
fig.suptitle('K-Means Image Compression',fontsize=20)
ax1.set_title('Compressed')
ax1.imshow(k_img)

ax2.set_title('Orignal (16,777,216 colors)')
ax2.imshow(img)
plt.subplots_adjust(top=0.85)
plt.show()
```
