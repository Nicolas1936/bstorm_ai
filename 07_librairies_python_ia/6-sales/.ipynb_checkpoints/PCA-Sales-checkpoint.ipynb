```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('6-Sales2.csv')
df.head()
```

```python
df.columns = df.columns.str.lower()
df.head()
```

## Data Cleaning

```python
# Suppression de la colonne "Unnamed: 0"
df.drop(columns=['unnamed: 0', 'week'], inplace=True)
```

```python
df
```

```python
# Mettre la colonne "qty" en nombre continue
df['qty'] = df['qty'].astype(int)
```

```python
df
```

```python
# 
df = pd.get_dummies(df)
df.head()
```

```python
df.drop('qty', axis=1, inplace=True)
```

```python
df.head()
```

```python
df = df.groupby('customer').sum()
```

## PCA

```python
pca_sales = PCA().fit(df)
```

```python
pca_sales.explained_variance_
```

```python
pca_sales.explained_variance_ratio_
```

```python
# variance expliquÃ©e
plt.bar(x=range(1, 17), height=pca_sales.explained_variance_, color="red")
plt.xlabel("Nombre de composantes")
plt.ylabel("Variance expliquÃ©e")
plt.plot(range(1, 17), pca_sales.explained_variance_)
plt.show()

# ratio de la variance expliquÃ©e
plt.bar(x=range(1, 17), height=pca_sales.explained_variance_ratio_, color="green")
plt.xlabel("Nombre de composantes")
plt.ylabel("Ratio de la variance expliquÃ©e")
plt.plot(range(1, 17), pca_sales.explained_variance_ratio_)
plt.show()
```

```python
import numpy as np
```

```python
# variance expliquÃ©e
plt.bar(x=range(1, 17), height=pca_sales.explained_variance_, color="red")
plt.xlabel("Nombre de composantes")
plt.ylabel("Variance expliquÃ©e")
plt.plot(range(1, 17), pca_sales.explained_variance_)
plt.show()

# ratio de la variance expliquÃ©e
plt.bar(x=range(1, 17), height=np.cumsum(pca_sales.explained_variance_ratio_), color="green")
plt.xlabel("Ratio du nombre de composantes")
plt.ylabel("Cumul du ratio de la variance expliquÃ©e")
plt.show()
```

```python
# on affiche l'apport de chaque feature dans chaque composante
for i in range(3):
    print(pca_sales.components_[i])
```

```python
pca_sales.components_.shape
```

```python
for i in range(3):
    plt.bar(df.columns, pca_sales.components_[i])
    plt.xticks(rotation= 90)
    plt.title('CP' + str(i+1) + ' Variance expliquÃ©e ' + str(round(pca_sales.explained_variance_ratio_[i] * 100, 2)) + "%")
    plt.show()
```

```python
for i in range(3):
    plt.barh(df.columns, pca_sales.components_[i])
    #plt.xticks(rotation= 90)
    plt.title('CP' + str(i+1) + ' Variance expliquÃ©e ' + str(round(pca_sales.explained_variance_ratio_[i] * 100, 2)) + "%")
    plt.show()
```

```python
#PC1 : indicateur de personne ayant des enfants en bas age
#PC2 : indicateur de personnes sportifs
```

## Clustering

### Kmeans sur le dataset originel

```python
df_kmeans = df.drop(columns=['enfants', 'sportifs'])
```

```python
from sklearn.cluster import KMeans
```

```python
all_inertia = []

for i in range(1, 11):
    model_kmean = KMeans(n_clusters=i, random_state=0).fit(df_kmeans)
    all_inertia.append(model_kmean.inertia_)
```

```python
plt.plot(range(1, 11), all_inertia)
plt.xlabel('Nombres de clusters')
plt.ylabel('Cout du modÃ¨le (Inertia)')
plt.scatter(4, all_inertia[3], c='red')
```

### On constate que l'on obtient le mÃªme nombre de clusters => 4
### Affichons les labels correspondants aux customers afin de les comparer avec les groupements gÃ©nÃ©rÃ©s grÃ¢ce Ã  la PCA

```python
model_kmean_4 = KMeans(n_clusters=4, random_state=0).fit(df_kmeans)
```

```python
model_kmean_4.labels_
```

```python
df_kmeans['labels'] = model_kmean_4.labels_
```

```python
df_kmeans
```

```python
df['labels'] = model_kmean_4.labels_
df
```

### DÃ©rivons des 2 colonnes Enfants et Sportifs, une colonne de label propre Ã  la PCA

```python
df.loc[(df['enfants'] == False) & (df['sportifs'] == False), 'label PCA'] = 2
df.loc[(df['enfants'] == True) & (df['sportifs'] == False), 'label PCA'] = 0
df.loc[(df['enfants'] == False) & (df['sportifs'] == True), 'label PCA'] = 1
df.loc[(df['enfants'] == True) & (df['sportifs'] == True), 'label PCA'] = 3
```

```python
df['label PCA'] = df['label PCA'].astype(int)
```

```python
df = df.rename(columns={'labels' : 'label KMeans'})
```
