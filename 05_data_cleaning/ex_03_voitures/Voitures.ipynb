### engine size

```python
df.head(15)
```

```python
df['engine size2'] = df['engine size2'].fillna(value=df['engine size'])
```

```python
df.drop(columns='engine size', inplace=True)
```

```python
df.columns = df.columns.str.replace('engine size2', 'engine size')
```

```python
df = df[df['engine size'] != 'Unknown']
```

```python
df['engine size'].unique()
```

```python
df['engine size'] = df['engine size'].str.replace(' cc', '').str.replace('T', '').str.replace(',', '.').copy()
```

```python
#df['engine size'] = df['engine size'].map(lambda x : x.replace(' cc', '')).copy()
```

```python
df['engine size'].unique()
```

```python
#df['engine size'] = round(df['engine size'].astype(float), 1).copy()
df['engine size'] = df['engine size'].astype(float).copy()
```

```python
def myfunc(x):
    if x >3.0:
        return x / 1000
    return x

df['engine size'] = df['engine size'].map(myfunc)
```
