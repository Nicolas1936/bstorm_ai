```python
import pandas as pd
import numpy as np

df = pd.read_csv('data/unclean cclass.csv')
df['model'] = 'C Class'
df.drop(columns='reference', inplace=True)
colomns = df.columns[1:]
df = df.dropna(axis=0, how='all', subset=colomns)
df = df.dropna(subset=['year'])
df = df.copy()
df['price'] = df['price'].str.replace('£' , '').copy()
df['price'] = df['price'].str.replace(',' , '')
df['price'] = df['price'].str.strip()
df['price'] = df['price'].astype(int)

def myfunc(x):
    if x in ['Petrol', 'Diesel']:
        return x
    return np.nan

df['fuel type'] = df['fuel type'].map(myfunc).copy()
df['fuel type'] = df['fuel type'].fillna(df['fuel type2'])
df.drop(columns='fuel type2', inplace=True)
df['engine size2'] = df['engine size2'].fillna(value=df['engine size'])
df.drop(columns='engine size', inplace=True)
df.columns = df.columns.str.replace('engine size2', 'engine size')
df = df[df['engine size'] != 'Unknown']
df['engine size'] = df['engine size'].str.replace('CC', '').copy()
df['engine size'] = df['engine size'].astype(float).copy()
def myfunc(x):
    if x >3.0:
        return x / 1000
    return x

df['engine size'] = df['engine size'].map(myfunc)
df['engine size'] = round(df['engine size'], 1)
df['mileage'] = df['mileage'].fillna(df['mileage2'])
df.drop(columns='mileage2', inplace=True)
df['mileage'] = df['mileage'].str.replace(',', '')

#df['mileage'] = df['mileage'].astype(int)


df
```
