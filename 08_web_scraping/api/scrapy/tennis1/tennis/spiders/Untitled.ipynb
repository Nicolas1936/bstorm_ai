```python
import pandas as pd


header_list = ['id', 'firstname', 'lastname', 'hand', 'date', 'country']
data = pd.read_csv('https://raw.githubusercontent.com/JeffSackmann/tennis_atp/master/atp_players.csv', 
                    names=header_list)
data.to_csv('data.csv', index=False)
```

```python
data = pd.read_csv('data.csv')
data
```

```python
import requests as rq
import time

def search_url(firstname, lastname):
    time.sleep(0.5)
    r = rq.get(f'https://en.wikipedia.org/w/api.php?action=opensearch&format=json&formatversion=2&search={firstname} {lastname}')
    dict_ = r.json()
    url = dict_[3][0]
    return url

for index, row in data.head(20).iterrows():
    data.at[index, 'wiki_url'] = search_url(row['firstname'], row['lastname'])
```

```python
data.to_csv('data_with_url.csv', index=False)
```
