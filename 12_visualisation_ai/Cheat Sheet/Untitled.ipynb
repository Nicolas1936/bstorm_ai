```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.relplot(data=tips,
            x='total_bill',
            y='tip')
```

```python
sns.relplot(data=tips,
            x='total_bill',
            y='tip',
            hue='time')
```

```python
sns.relplot(data=tips,
            x='total_bill',
            y='tip',
            hue='time',
            style='smoker')
```

```python
sns.relplot(data=tips,
            x='total_bill',
            y='tip',
            hue='size',
            style='time',
            col='smoker')
```
