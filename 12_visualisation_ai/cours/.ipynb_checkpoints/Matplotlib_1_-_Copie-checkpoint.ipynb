### Plot (x,y)

La fonction __plot()__ est utilisée pour dessiner des points (marqueurs) dans un espace à 2 dimensions.

Par défaut, la fonction plot () trace une ligne d'un point à l'autre.

La fonction plot () prend des paramètres pour spécifier des points.

    Le paramètre 1 est un tableau contenant les points sur l'axe des x.

    Le paramètre 2 est un tableau contenant les points sur l'axe y.

Argument rapide "go" (green et rond) afin de préciser si l'on veut juste les points à défaut d'une courbe reliant les points (<> -b).


__<font color="green">- Ajout de paramètres supplémentaires dans plot (x,y)</font>__


    1) __Marqueurs__

Marqueurs (exemple: marker = 'o')
Autres: '*'; '+'; 's'; '.'; 'd'; 'h'; '<'; '_'; 'p'

Taille des marqueurs ( exemple: ms= 10)

Couleur du bord du marqueur (exemple : mec = 'r' / mec='#4CAF50')

Couleur du marqueur (exempel: mfc = 'r' / mfc='#4CAF50')

    2) __Ligne__

Style de la Ligne/Courbe (exemple: linestyle = 'dotted' / ls= ':')
Autres: '-'; '--','-.'

Couleur du trait (exemple; color='r' / c = '#4CAF50')

Epaisseur du trait (exemple: linewidth = '20.5'/lw='10')

    3) __Nom du trait (label="Nom")__

Ce paramètre est utile si l'on désire afficher une légende sur le graphique (voir plus bas)

__<font color="green"> - Ajout de "label" et "title"</font>__

Il est possible d'ajouter les fonctions xlabel () et ylabel () pour définir une étiquette pour les axes x et y.

Il est possible d'ajouter la fonction title () pour définir un titre pour le tracé. 

__<font color="green">- Propriétés avec fontdict()</font>__

Il est possible d'utiliser le paramètre fontdict dans xlabel (), ylabel () et title () pour définir les propriétés de police du titre et des étiquettes. 

__<font color="green">- Propriétés: loc()</font>__
  
Permet de positionner le titre ou legende


__<font color="green">- Ajout de lignes/courbes supplémentaires</font>__

Pour dessiner une ligne supplementaire sur le graphe, il suffit d'ajouter la fonction __plot()__ avec les nouvelles données

__<font color="green">- Ajout d'une légende et changer l'échelle des axes x et y </font>__

Il est possible d'ajouter une légende sur un graphique au moyen de la fonction plt.legend(). 

Cette fonction nécessite de préciser le label dans la fonction plt.plot().

Les fonctions plt.xlim et plt.ylim définissent les plages limites pour les axes. P

ar exemple, plt.xlim ([- 50, 50]) limite le tracé pour n'afficher que les valeurs x comprises entre -50 et 50.


__<font color="green">- Ajouter un quadrillage </font>__

 La fonction grid () permet d'ajouter des lignes de quadrillage au tracé. 
 
 * Propriétés: Précise si juste sur un axe particulier - (axis = 'y') (axis = 'x')
 
Même propriété que le plot()

--> color = 'green', linestyle = '--', linewidth = 0.5

__<font color="green">- Taile du graphique  </font>__


__<font color="green">- Sauver un graphique  </font>__

Il est possible de sauver le graphique dans le répertoire de travail  (ou un autre endroit si on précise le chemin) sous un format d'image comme jpeg ou png. 
La fonction plt.savefig() permettra de faire cela.

### Bar plot

Les graphiques à barres sont l'un des types de graphiques les plus courants et sont utilisés pour afficher les données associées aux variables catégorielles. 

Pyplot fournit une méthode __bar()__ pour créer des graphiques à barres qui prennent des arguments : des variables catégorielles, leurs valeurs et leur couleur (si vous souhaitez en spécifier). 

__<font color='green'> - Graphe en barres horizontales </font>__

Pour créer des graphiques à barres horizontales, utilisez la méthode __barh()__. 

__<font color='green'> - Argument d'erreur </font>__

Nous pouvons également passer un argument (avec sa valeur) __xerr  ou yerr__ pour décrire la variance/Ecart-type de nos données

__<font color='green'> - Empilement horizontal et vertical </font>__

Pour créer des graphiques à barres empilés horizontalement, nous utilisons deux fois la méthode bar() et passons les arguments où nous mentionnons l'index et la largeur de nos graphiques à barres afin de les empiler horizontalement. 

Pour empiler verticalement les graphiques à barres, vous pouvez utiliser un argument __bottom__ et mentionner le graphique à barres que vous voulez empiler en dessous comme valeur. 

legend() est utilisé pour afficher la légende du graphique et xticks() pour étiqueter notre axe des x en fonction de la position de des barres

```python
# Création d'un dataframe
df=pd.DataFrame(np.random.rand(50,1)*100, columns=['vitesse'])
df['poids']=df['vitesse']**2
df['Nbre_accident']=np.random.randint(10, size=50)
df['category']=np.random.choice(['Ford','BMW', 'Honda','Toyota'], size=50)
df['color']=np.random.choice(['Bleu','Rouge', 'Noir'], size=50)
df.head()
```

```python
# Exemple 2 dataframe - ERROR
plt.bar(df.category,df.Nbre_accident, color = "red",width = 0.50)
plt.show()
```

```python
# Exemple 2 dataframe - SOLUTION
x=df.groupby('category')['Nbre_accident'].sum().reset_index().category
y=df.groupby('category')['Nbre_accident'].sum().reset_index().Nbre_accident

plt.bar(x,y, color = "green",width = 0.80) # width precise l'epaisseur de barre
plt.show()
```

#### Graphe en barres horizontales

```python
x=df.groupby('category')['Nbre_accident'].sum().reset_index().category
y=df.groupby('category')['Nbre_accident'].sum().reset_index().Nbre_accident

plt.barh(x,y,color = "green",height = 0.3) # height precise l'epaisseur de barre
plt.show()
```

#### Graphe en barres avec des barres d'error afin de donner la dispersion des données

```python
x=df.groupby('category')['Nbre_accident'].mean().reset_index().category
y=df.groupby('category')['Nbre_accident'].mean().reset_index().Nbre_accident
z=df.groupby('category')['Nbre_accident'].std() # standard deviation - ecart type
plt.bar(x,y,color = "green", yerr=z)
plt.show()
```

```python
x=df.groupby('category')['Nbre_accident'].sum().reset_index().category
y=df.groupby('category')['Nbre_accident'].sum().reset_index().Nbre_accident
z=df.groupby('category')['Nbre_accident'].std() # standard deviation - ecart type
plt.bar(x,y,color = "green", yerr=z)
plt.xticks(rotation=45) # rotation des labels x
```

```python
x=df.groupby('category')['Nbre_accident'].sum().reset_index().category
y=df.groupby('category')['Nbre_accident'].sum().reset_index().Nbre_accident
z=df.groupby('category')['Nbre_accident'].std() # standard deviation - ecart type
plt.barh(x,y,color = "green", xerr=z)
plt.show()
```

#### Empilement 

```python
# Empilement vertical
x=df.groupby('category')['Nbre_accident'].sum().reset_index().category
y=df.groupby('category')['Nbre_accident'].sum().reset_index().Nbre_accident
z=df.groupby('category')['Nbre_accident'].count().reset_index().Nbre_accident
index=np.arange(4)

plt.bar(x,y, color = "green",width = 0.30) # width precise l'epaisseur de barre
plt.bar(x,z, color = "blue",width = 0.30
        , bottom=y) # bottom permet d'empiler aux 1eres valeurs
plt.bar(x,z, color = "red",width = 0.30
        , bottom=z)

plt.show()
```

```python
# Empilement horizontal
x=df.groupby('category')['Nbre_accident'].sum().reset_index().category
y=df.groupby('category')['Nbre_accident'].sum().reset_index().Nbre_accident
z=df.groupby('category')['Nbre_accident'].count().reset_index().Nbre_accident
index=np.arange(4)

plt.bar(x,y, color = "green",width = 0.30) # width precise l'epaisseur de barre
plt.bar(index+0.3,z, color = "blue",width = 0.30)
plt.show()
```

### Pie Chart

Un autre type de graphique de base est un graphique à secteurs utilisé pour les variables catégorielles.
Il peut être créé à l'aide de la méthode __pie ()__

Il est possible de transmettre des arguments pour personnaliser le graphique à secteurs pour afficher l'ombre, en exploser une partie, l'incliner selon un angle comme suit 

```python
df.groupby('category')['category'].count()
```

```python
# Exemple 2 dataframe
x=df.groupby('category')['Nbre_accident'].count().reset_index().category
y=df.groupby('category')['category'].count()

plt.figure(figsize=(8,8))
p, tx, autotexts=plt.pie(y, labels = x, autopct='%1.2f%%')

# Pour afficher simplement les nombres bruts
#for i, a in enumerate(autotexts):
#    a.set_text("{}".format(y[i]))
# 
plt.show() 
```

```python
# Exemple 2 avec une légende
x=df.groupby('category')['Nbre_accident'].count().reset_index().category
y=df.groupby('category')['category'].count()

plt.figure(figsize=(8,8))
plt.pie(y, labels = x, autopct='%1.2f%%')
plt.legend(loc="upper right") #
plt.show()
```

```python
# Exemple 2 - Explode 
x=df.groupby('category')['Nbre_accident'].count().reset_index().category
y=df.groupby('category')['category'].count()

colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue'] # spécifie les couleurs aux labels
explode = (0, 0, 0.1, 0) # distance par rapport au centre

plt.pie(y, explode=explode, labels=x, colors=colors 
        ##,autopct='%1.1f%%'
       )
plt.figure(figsize=(8,8))
plt.show()
```

```python
# Exemple 2 - ombre et l'angle de départ
x=df.groupby('category')['Nbre_accident'].count().reset_index().category
y=df.groupby('category')['category'].count()

colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue'] # spécifie les couleurs aux labels
explode = (0.1, 0, 0, 0) # distance par rapport au centre

plt.pie(y, explode=explode, labels=x, colors=colors,autopct='%1.1f%%', shadow=True, startangle=140)
plt.figure(figsize=(8,8))
plt.show()
```

### Histogramme

Les histogrammes sont un type de tracé très courant lorsque nous examinons des variables continues telles que la taille et le poids, les cours des actions, le temps d'attente pour un client, etc. 

Les données de l'histogramme sont tracées dans une plage par rapport à sa fréquence. 

Les histogrammes sont des graphiques très courants dans les probabilités et les statistiques et constituent la base de diverses distributions telles que la distribution normale, la distribution t, etc. 

Bins = the range of values

```python
# Exemple
plt.hist(df.vitesse, bins=20)
plt.title("Histogramme")
plt.xlabel("variable vitesse")
plt.ylabel("Frenquence")
plt.show()
```

#### Annotation

*   `s`: str, the text of annotation.
*   `xy`: Tuple specifying the (x,y) point to annotate (in this case, end point of arrow).
*   `xytext`: Tuple specifying the (x,y) point to place the text (in this case, start point of arrow).
*   `xycoords`: The coordinate system that xy is given in - 'data' uses the coordinate system of the object being annotated (default).
*   `arrowprops`: Takes a dictionary of properties to draw the arrow:
    *   `arrowstyle`: Specifies the arrow style, `'->'` is standard arrow.
    *   `connectionstyle`: Specifies the connection type. `arc3` is a straight line.
    *   `color`: Specifies color of arrow.
    *   `lw`: Specifies the line width.

http://matplotlib.orsg/api/pyplot_api.html#matplotlib.pyplot.annotate.


```python
plt.figure(figsize=(12,8))
plt.scatter(df.vitesse, df.poids) 

plt.xlabel('Variable x')
plt.ylabel('Variable y')
plt.title('Ceci est un etitre')

# Annotate arrow
plt.annotate('blblblbl',  # s: str. Will leave it blank for no text
             xy=(45, 70),  # fleche termine x, termine y
             xytext=(28, 100),  # text commence x - commence y 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2)
             )

plt.show()
```

### Multi-Plot (x,y)

Nous pouvons utiliser la méthode __subplot()__ pour ajouter plusieurs tracés dans une figure. 

On peut utiliser cette méthode pour séparer deux graphiques que nous avons tracés sur les mêmes axes. 

La méthode subplot() prend trois arguments : 
    
    * nrows (nombre de lignes)
   
    * ncols (nombre de colonnes)
    
    * index (le numéro d'index)  

Par exemple, nous voulons créer deux sous-graphique dans une figure de telle sorte qu'elle se présente sur une ligne et sur deux colonnes et nous passons donc les arguments (1,2,1) et (1,2,2) dans la méthode subplot(). 

Notez que vous pouvez utiliser séparément la méthode title() pour les deux sous-graphe. 

Vous pouvez également utiliser la méthode suptitle() pour créer un titre centralisé pour la figure. 

```python
# Exemple 1
x= np.linspace(0,2,10)
y= x**2

plt.figure(figsize=(12,8)) # Taille de la figure

plt.subplot(2,1,1) # Sous graphe ligne 2, colonne 1, index 1
plt.plot(x,y,c='red') # un graphe sur le 1er sous graphe
plt.title('Titre')

plt.subplot(2,1,2) # Sous graphe ligne 2, colonne 1, index 2
plt.plot(x,np.sin(x),label='sin') # un 1er graphe sur le 2e sous graphe
plt.plot(x,np.cos(x),label='cos') # un 2e graphe sur le 2e sous graphe
plt.legend()
```

### Histo - Scatter // jitter plot seaborn 

#### Exemple 1


```python
np.random.seed(19680801)

# some random data
x = np.random.randn(1000)
y = np.random.randn(1000)

def scatter_hist(x, y, ax, ax_histx, ax_histy):
    # no labels
    ax_histx.tick_params(axis="x", labelbottom=False)
    ax_histy.tick_params(axis="y", labelleft=False)

    # the scatter plot:
    ax.scatter(x, y)

    # now determine nice limits by hand:
    binwidth = 0.25
    xymax = max(np.max(np.abs(x)), np.max(np.abs(y)))
    lim = (int(xymax/binwidth) + 1) * binwidth

    bins = np.arange(-lim, lim + binwidth, binwidth)
    ax_histx.hist(x, bins=bins)
    ax_histy.hist(y, bins=bins, orientation='horizontal')

# definitions for the axes
left, width = 0.1, 0.65
bottom, height = 0.1, 0.65
spacing = 0.005


rect_scatter = [left, bottom, width, height]
rect_histx = [left, bottom + height + spacing, width, 0.2]
rect_histy = [left + width + spacing, bottom, 0.2, height]

# start with a square Figure
fig = plt.figure(figsize=(8, 8))

ax = fig.add_axes(rect_scatter)
ax_histx = fig.add_axes(rect_histx, sharex=ax)
ax_histy = fig.add_axes(rect_histy, sharey=ax)

# use the previously defined function
scatter_hist(x, y, ax, ax_histx, ax_histy)

plt.show()

```
